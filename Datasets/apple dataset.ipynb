## Crea un decision tree per classificar la qualitat de les pomes i identifica les millors característiques per determinar si una poma és bona o no.

Imports:

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

1. Carregar el csv

```python
apple = pd.read_csv('apple_quality.csv')

print(apple.head())
```

2. Definir la variable objectiu que en aquest cas serà la columna "Quality" i la variable random_state

```python
X = apple.drop(columns=["A_id", "Quality"])
y = apple["Quality"]

rs = 42
```

3. Transformar el contingut de la columna "Quality" a valors booleans

```python
y = LabelEncoder().fit_transform(y)
print(y)
```

4. Dividir les dades en entrenament i prova

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=rs)
```

5. Crear i entrenar el decision tree

```python
classifier = DecisionTreeClassifier(random_state=rs)
classifier.fit(X_train, y_train)
```

6. Predir i calcular la precisió

```python
y_pred = classifier.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("Precisió del decision tree: ", acc)
```

7. Identificar les característiques més rellevants

```python
feature_importances = pd.Series(classifier.feature_importances_, index=X.columns)
print("Importància de les característiques:")
print(feature_importances.sort_values(ascending=False))
```
